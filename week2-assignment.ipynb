```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
df=sns.load_dataset('car_crashes')
df
```

```python
print(df.info())
df.head()
```

```python
sns.scatterplot(x="total",y="speeding",data=df)
# --> as percentage of car crashes increases, percentage of car crashes due to speeding increases -- positive correlation
```

```python
sns.scatterplot(x="total",y="alcohol",data=df)

# --> as percentage of car crashes increases, percentage of car crashes due to alcohol increases -- positive correlation
```

```python
sns.scatterplot(x="total",y="not_distracted",data=df)
# --> Always a high percentage of car crashes are with people who are not distracted and
# --> as percentage of car crashes increases, percentage of car crashes with people not distracted increases -- positive correlation
```

```python
sns.scatterplot(x="total",y="no_previous",data=df)

# --> as percentage of car crashes increases, percentage of car crashes with people who had no previous accidents increases -- positive correlation
```

```python
sns.scatterplot(x="total",y="ins_losses",data=df)

# --> no correlation between total accidents and insurance losses 
```

```python
sns.scatterplot(x="total",y="ins_premium",data=df)

# --> no correlation between total accidents and insurance premium 
```

```python
sns.scatterplot(x="alcohol",y="speeding",data=df)

# --> positive correlation between alcohol and speeding
```

```python
sns.scatterplot(x="speeding",y="no_previous",data=df)

# --> no correlation between speeding and no previous accidents 
```

```python
sns.lineplot(x="total",y="speeding",data=df,ci=None) #blue
sns.lineplot(x="total",y="alcohol",data=df,ci=None) #orange
sns.lineplot(x="total",y="not_distracted",data=df,ci=None) #green
sns.lineplot(x="total",y="no_previous",data=df,ci=None) #red
```

```python
sns.distplot(df["total"])

#total accidents are dense in the 10-20% range for all states
```

```python
sns.distplot(df["speeding"])

#speeding percentage are dense in the 2-6% range for all states
```

```python
sns.distplot(df["alcohol"])

#alcohol caused crashes percentage are dense in the 3-6% range for all states
```

```python
sns.distplot(df["not_distracted"])

#people who arent distracted percentage are dense in the 13-18% range for all states
```

```python
sns.distplot(df["no_previous"])

#accidents with people who had no previous ones percentage are dense in the 12-18% range for all states
```

```python
sns.relplot(x="total",y="speeding",data=df,hue="alcohol") 
```

```python
sns.relplot(x="alcohol",y="speeding",data=df,hue="not_distracted") 
# if percentage of speeding and alcohol related crashes are less, the percentage of crashes due to driver not distracted is less too
# have positive correlation
```

```python
sns.relplot(x="alcohol",y="speeding",data=df,hue="no_previous") 
# if percentage of speeding and alcohol related crashes are less, the percentage of crashes with driver with no previous is less too
# it is positive correlation with alcohol and speeding
```

```python
sns.jointplot(x="total",y="speeding",data=df) 
#combination of a bar plot and scatter plot, it has positive correlation,
# and total is dense in the 8-12% range and speeding is dense in the 4-5% range 
```

```python
sns.boxplot(x="total",y="speeding",data=df) #it is very large because there are many y values
```
